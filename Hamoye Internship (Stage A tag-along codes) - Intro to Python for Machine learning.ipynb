```python
# checking for the total number and percentage of missing data in the year 2014 
missing_data = food_df['Y2014'].isnull().sum()
total_data = food_df['Y2014'].shape[0]
percentage_missing = (missing_data / total_data) * 100
print(f"Total number of missing data in column Y2014: {missing_data}")
print(f"Percentage of missing data in column Y2014: {percentage_missing:.3f}%")
```
